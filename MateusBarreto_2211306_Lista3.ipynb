```python
import pandas as pd
import json
import random
import numpy as np
from google.colab import drive



drive.mount('/drive')

arq1 = pd.read_csv("/drive/MyDrive/ENG1111/class_titanic.csv")
display(arq1)

#Escolhendo uma coluna (Age) e apagando as linhas com valores nulos
arq1.dropna(subset=["Age"],inplace=True)
display(arq1)

#Calculando a média da mesma coluna (Age)
media_idade=arq1["Age"].mean()
print("\nMedia de idade: %.2f anos\n"%media_idade)

#Eliminando linhas que possuam valores muito distantes da média (25 para baixo e 30 para cima)
#E criando um novo dataframe
arq2 = arq1[(media_idade - 25 <= arq1['Age']) & (arq1["Age"] <= media_idade + 30)]
display(arq2)

#Criando um novo Dataframe que contenha apenas 5 colunas do conjunto anterior
arq3 = arq2[["PassengerId","Name","Sex","Age","Survived"]]
display(arq3)

#Criando uma nova coluna que siga a regra: (Coluna A - Coluna B)*número aleatório (entre 0 e 1)
arq3.insert(5,'NewColumn',(arq3['PassengerId']-arq3['Age'])*random.random())
display(arq3)

#Embaralhando o dataframe
arq3=arq3.reindex(np.random.permutation(arq3.index))
display(arq3)

#Ordenando por ordem crescente de idade
arq3 = arq3.sort_values(by='Age',ascending=True)
display(arq3)

#Usando a função groupby para idade (Age) e para sexo (Sex)
idade = arq3.groupby('Age')[['Age']].count()
sexo = arq3.groupby('Sex')[['Sex']].count()
display(idade)
display(sexo)

#Usando a função MAPS para substituir o sexo feminino para 0 e masculino para 1
arq3["Sex"] = arq3["Sex"].map(
    {'female':0, 'male':1 })
display(arq3)

#
arq3.to_csv("/drive/MyDrive/ENG1111/novo_arquivo.csv")
arq3.to_json("/drive/MyDrive/ENG1111/novo_arquivo.json")
```
